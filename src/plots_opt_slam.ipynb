# Static Results

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('../results/slam_static_optimization_results.csv')

# Set up the figure and subplots
fig, axs = plt.subplots(5, 1, figsize=(10, 12), sharex=True)

# Define highlight color for the best Q/R value
highlight_color = 'red'

# Scatter plot for X
axs[0].scatter(df['ATE_MSRE'], df['X']**2, color='blue')
axs[0].scatter(df['ATE_MSRE'][0], df['X'][0]**2, color=highlight_color, label='Best Q/R value', zorder=5)
axs[0].set_ylabel('X [m]')
axs[0].grid(True)
axs[0].legend()

# Scatter plot for Y
axs[1].scatter(df['ATE_MSRE'], df['Y']**2, color='green')
axs[1].scatter(df['ATE_MSRE'][0], df['Y'][0]**2, color=highlight_color, zorder=5)
axs[1].set_ylabel('Y [m]')
axs[1].grid(True)

# Scatter plot for Theta
axs[2].scatter(df['ATE_MSRE'], np.degrees(df['Theta']**2), color='orange')
axs[2].scatter(df['ATE_MSRE'][0], np.degrees(df['Theta'][0]**2), color=highlight_color, zorder=5)
axs[2].set_ylabel('Heading Angle [degrees]')
axs[2].grid(True)

# Scatter plot for Range
axs[3].scatter(df['ATE_MSRE'], df['Range']**2, color='purple')
axs[3].scatter(df['ATE_MSRE'][0], df['Range'][0]**2, color=highlight_color, zorder=5)
axs[3].set_ylabel('Range [m]')
axs[3].grid(True)

# Scatter plot for Bearing
axs[4].scatter(df['ATE_MSRE'], df['Bearing']**2, color='brown')
axs[4].scatter(df['ATE_MSRE'][0], df['Bearing'][0]**2, color=highlight_color, zorder=5)
axs[4].set_ylabel('Bearing [degrees]')
axs[4].set_xlabel('ATE MSRE')
axs[4].grid(True)

# Set the main title
fig.suptitle('Q and R Variables vs ATE MSRE', fontsize=16)
axs[0].set_xlim(1.0, 25)
plt.tight_layout()

# Show the plot
plt.show()

```
